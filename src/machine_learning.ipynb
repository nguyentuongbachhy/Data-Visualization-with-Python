```python
import pandas as pd
import numpy as np
```

```python
confirmed_df = pd.read_csv("../data/raw/time_series_covid_19_confirmed.csv")
deaths_df = pd.read_csv("../data/raw/time_series_covid_19_deaths.csv")
recovered_df = pd.read_csv("../data/raw/time_series_covid_19_recovered.csv")
```

```python
print("So luong du lieu:")
print("Confirmed: ", confirmed_df.shape)
print("Deaths: ", deaths_df.shape)
print("Recovered: ", recovered_df.shape)
```

Khac nhau giua so luong quoc gia co so lieu => Can loai tru cac dong khong co so lieu ve Recovered

```python
place_name_recovered = [f"{lat_value}_{long_value}" for lat_value, long_value in zip(recovered_df["Province/State"], recovered_df["Country/Region"])]
place_name_deaths = [f"{lat_value}_{long_value}" for lat_value, long_value in zip(deaths_df["Province/State"], deaths_df["Country/Region"])]
place_name_confirmed = [f"{lat_value}_{long_value}" for lat_value, long_value in zip(confirmed_df["Province/State"], confirmed_df["Country/Region"])]

confirmed_df["Province/State/Country/Region"] = place_name_confirmed
deaths_df["Province/State/Country/Region"] = place_name_deaths
recovered_df["Province/State/Country/Region"] = place_name_recovered

drop_idx = confirmed_df[~confirmed_df["Province/State/Country/Region"].isin(place_name_recovered)].index
confirmed_df.drop(drop_idx, inplace=True)
confirmed_df.reset_index(inplace=True)
drop_idx = deaths_df[~deaths_df["Province/State/Country/Region"].isin(place_name_recovered)].index
deaths_df.drop(drop_idx, inplace=True)
deaths_df.reset_index(inplace=True)
drop_idx = recovered_df[~recovered_df["Province/State/Country/Region"].isin(place_name_confirmed)].index
recovered_df.drop(drop_idx, inplace=True)
recovered_df.reset_index(inplace=True)
recovered_df.shape
```

```python
confirmed_data = confirmed_df.drop(columns=["Province/State", "Country/Region", "Lat", "Long", "index", "Province/State/Country/Region"])
deaths_data = deaths_df.drop(columns=["Province/State", "Country/Region", "Lat", "Long", "index", "Province/State/Country/Region"])
recovered_data = recovered_df.drop(columns=["Province/State", "Country/Region", "Lat", "Long", "index", "Province/State/Country/Region"])

data = []

for i in range(0,confirmed_data.shape[0]):
    cf_row = confirmed_data.iloc[i]
    de_row = deaths_data.iloc[i]
    rec_row = recovered_data.iloc[i]
    new_row = pd.concat([cf_row, de_row, rec_row], axis=1)
    data.append(new_row)
data = np.array(data)
data.shape
```
